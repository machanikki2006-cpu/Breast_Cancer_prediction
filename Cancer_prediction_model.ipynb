```python
import pandas as pd
import numpy as np
import matplotlib as mp
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix,precision_score,recall_score,f1_score
```

```python
df=pd.read_csv(r"C:\Users\M.Nikhileswara\OneDrive\Desktop\Machine Learning\Cancer_prediction\breast-cancer.csv")
df.shape
```

```python
df=df.drop(['id'],axis=1)
df.head()
```

```python
dummy=pd.get_dummies(df['diagnosis'],drop_first=True,dtype=int)
```

```python
dummy.value_counts()
```

```python
df=df.drop(['diagnosis'],axis=1)
```

```python
df=pd.concat([df,dummy],axis=1)
```

```python
df.head()
```

```python
df.info()
X=df.drop(['M'],axis=1)
y=df['M']
```

```python
scaler=StandardScaler()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
X_trainsc=scaler.fit_transform(X_train)
X_testsc=scaler.transform(X_test)
```

```python
log=LogisticRegression(class_weight='balanced')
log.fit(X_trainsc,y_train)
```

```python
y_logpred=log.predict(X_testsc)
```

```python
print(classification_report(y_test,y_logpred))
```

```python
print('precision_score:',precision_score(y_test,y_logpred))
print('recall_score:',recall_score(y_test,y_logpred))
print('f1_score:',f1_score(y_test,y_logpred))
print(confusion_matrix(y_test,y_logpred))
```

```python
print(classification_report(y_test,y_logpred))
```

```python
svm=SVC()
param_grid = {
    'C': [1,6,7,8,9,10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto',1, 0.1, 0.01]
}
```

```python
grid = GridSearchCV(
    estimator=svm,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
)
```

```python
grid.fit(X_trainsc, y_train)
```

```python
print(grid.best_params_)
```

```python
y_svmpred=grid.predict(X_testsc)
```

```python
print('precision_score:',precision_score(y_test,y_svmpred))
print('recall_score:',recall_score(y_test,y_svmpred))
print('f1_score:',f1_score(y_test,y_svmpred))
print(confusion_matrix(y_test,y_svmpred))
```

```python
print(classification_report(y_test,y_svmpred))
```
